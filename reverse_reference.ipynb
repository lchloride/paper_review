```python
import pandas as pd
import numpy as np
import json
import traceback
```

```python
train=pd.read_csv("./data/mag_reduced_0.csv", dtype={'field':str, 'venue':str})
```

```python
train = train.set_index(['id'])
train.references = train.references.fillna('[]')
train.head()
```

```python
rev_ref_dict = {}
id_list = train.index.values.tolist()
for idd in id_list:
    rev_ref_dict[idd] = [[]]
```

```python
cnt = 0
def setup_rev_ref(row):
#     print(row.name)
    global cnt
    if cnt % 100000 == 0:
        print(cnt)
    cnt += 1
    
    ref_str = row['references']
    row_id = row.name
    ref_list = json.loads(ref_str)
    
    for index, ref in enumerate(ref_list):
        try:
            rev_ref_dict[ref][0].append(row_id)
        except KeyError as e:
            rev_ref_dict[ref] = [[row_id]]
            
    return row
```

```python
train.apply(setup_rev_ref, axis=1)
```

```python
count = 0
for key in rev_ref_dict:
    if len(rev_ref_dict[key]):
        count += 1
count
```

```python
for idx, idd in enumerate(rev_ref_dict):
    rev_ref_dict[idd].append(len(json.loads(train.loc[idd].references)))
```

```python
len(rev_ref_dict)
```

```python
rev_ref_df = pd.DataFrame.from_dict(rev_ref_dict, orient='index')
```

```python
rev_ref_df.rename(columns=dict(zip(rev_ref_df.columns[[0,1]], ['cited', 'n_ref'])),inplace=True)
rev_ref_df.head()
```

```python
rev_ref_df['cited'] = rev_ref_df['cited'].apply(json.dumps)
```

```python
rev_ref_df.to_csv('./data/rev_def.csv', encoding='utf-8')
```
