```python
import numpy as np
import pandas as pd
import json
```

```python
with open('./data/author_score_3.json', 'r') as f:
    author_score = json.load(f, encoding='utf-8')
```

```python
for author in author_score:
    print(author, author_score[author])
    break
```

```python
with open('./data/mag_list_noid.json', 'r') as f:
    mag_list = json.load(f, encoding='utf-8')
```

```python
for author in author_score:
    row = author_score[author]
    n_citation = []
    for iid in row['id']:
        n_citation.append(mag_list[iid]['n_citation'])
    author_score[author]['n_cits'] = n_citation
```

### author in  computer science or physics

```python
for author in author_score:
    row = author_score[author]
    cs_cnt = 0
    phy_cnt = 0
    for iid in row['id']:
        if mag_list[iid]['field'] == 'Computer Science':
            cs_cnt += 1
        elif mag_list[iid]['field'] == 'Physics':
            phy_cnt += 1
    if cs_cnt >= phy_cnt:
        author_score[author]['major'] = 'Computer Science'
    else:
        author_score[author]['major'] = 'Physics'
```

```python
with open('./data/author_score_hindex_3.json', 'w') as f:
    json.dump(author_score, f, ensure_ascii=False)
```

### Calc h_index

```python
with open('./data/author_score_hindex_3.json', 'r') as f:
    author_score = json.load(f, encoding='utf-8')
```

```python
for iid in author_score:
    print(author_score[iid])
    break
```

```python
def h_index(author_dic):
#     for key in author_dic:
#         author_dic[key]['h_index'] = 0
    for idx, key in enumerate(author_dic):
#         sorted(author_dic[key]['n_cits'], reverse=True)
        author_dic[key]['n_cits'].sort(reverse=True)
        author_dic[key]['h_index'] = max([min(k+1, v) for k,v in enumerate(author_dic[key]['n_cits'])]) if author_dic[key]['n_cits']!=[0] else 0
```

```python
h_index(author_score)
```

```python
with open('./data/author_score_hindex_3.json', 'w') as f:
    json.dump(author_score, f, ensure_ascii=False)
```
