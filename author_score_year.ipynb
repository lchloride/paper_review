```python
with open('./data_zx/author_score_3_final.json', 'r') as f:
    dic_au_score = json.load(f)
```

```python
for i, idx in enumerate(dic_au_score):
    if sum(dic_au_score[idx]['score_year']) != 0:
        print(dic_au_score[idx])
        break
```

```python
with open('./data_zx/keywords_coef.json', 'r') as f:
    kw_coef = json.load(f)
```

```python
print('total length:', len(dic_au_score))
for i, idx in enumerate(dic_au_score):
    row = dic_au_score[idx]
    row['score_year'] = [0] * 4
    for year in range(2014, 2018):
        for j in range(len(row['id'])):
            if dic_year[str(row['id'][j])]['year'] == year:
                row['score_year'][year-2014] += row['rank_score'][j]* 2**(-row['order'][j]) / kw_coef[row['id'][j]]
    if i % 1_000_000 == 0:
        print(i)
```
