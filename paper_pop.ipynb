```python
dic = dict((idx, {'cited': df.at[idx, 'cited'], 'year': 0}) for idx in df.index)
```

```python
# dic = df.to_dict(orient = 'index')
for i, idx in enumerate(dic):
    print(dic[idx])
    if i > 5:
        break
```

```python
with open('./data_zx/paper_year_dic', 'r') as f:
    dic_year = json.load(f)
```

```python
dic_year
```

```python
for i, idx in enumerate(dic):
    row = dic[idx]
    row['year'] = dic_year[str(idx)]['year']
    row['cited'] = row['cited'].replace(chr(39), chr(34))
```

```python
for i, idx in enumerate(dic):
    print(dic[idx])
    print(len(json.loads(dic[idx]['cited'])))
    if i > 5:
        break
```

```python
for i, idx in enumerate(dic):
    row = dic[idx]
#     row['cite_year'] = [0] * (2018-row['year']+1)
    for ref_id in json.loads(row['cited']):
        try:
            row['cite_year'][dic[ref_id]['year']-row['year']] += 1
        except Exception as e:
            pass
    if i % 2_000_000 == 0:
        print(i)
```

```python
total_cite = 0
for idx in dic.keys():
    total_cite += len(json.loads(dic[idx]['cited']))
print(total_cite)
```
